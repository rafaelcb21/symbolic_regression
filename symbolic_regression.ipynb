```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def build_graph(algorithm, title, y):
    df = pd.read_csv(algorithm)

    acc_n = df['n'].unique()
    soma_hits_por_n = df.groupby('n')['hits'].sum()
    soma_time_por_n = df.groupby('n')['time'].sum()

    # Criando uma figura com uma grade de subplots
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))

    # Plotando o primeiro gráfico no primeiro subplot
    axs[0].scatter(df['n'], df[y], s=0.5)
    axs[0].set_ylabel(y)
    axs[0].set_xlabel('n')
    axs[0].set_title(title + ' Algorithm')

    ## Plotando o segundo gráfico no terceiro subplot
    if y == 'time':
        axs[1].plot(acc_n, soma_time_por_n, linestyle='-')
        axs[1].set_ylabel(y)
        axs[1].set_xlabel('n')
        axs[1].set_title(title + ' Algorithm Accumulated')
    elif y == 'hits':
        axs[1].plot(acc_n, soma_hits_por_n, linestyle='-')
        axs[1].set_ylabel(y)
        axs[1].set_xlabel('n')
        axs[1].set_title(title + ' Algorithm Accumulated')

    # Plotando o terceiro gráfico no segundo subplot
    fig1 = plt.figure()
    axs1 = fig1.add_subplot(122, projection='3d')
    axs1.scatter(df['n'], df[y], df['code_line'], s=0.5)
    axs1.set_xlabel('n')
    axs1.set_ylabel(y)
    axs1.set_zlabel('code line')
    axs1.set_title(title + ' Algorithm')

    # Ajustando o layout
    plt.tight_layout()
    plt.show()
```

```python
print('*********** HITS ***********')

y_name = 'hits'
build_graph('study1.csv', 'Study 1', y_name)
build_graph('study2.csv', 'Study 2', y_name)
build_graph('shell.csv', 'Shell', y_name)
build_graph('selection_sort.csv', 'Selection Sort', y_name)
build_graph('quick_sort.csv', 'Quick Sort', y_name)
build_graph('heapsort.csv', 'Heapsort', y_name)
build_graph('insertion_sort.csv', 'Insertion Sort', y_name)
build_graph('insertion_sort_bin.csv', 'Insertion Sort Bin', y_name)
build_graph('binary_search.csv', 'Binary Search', y_name)
build_graph('binary_search_recursive.csv', 'Binary Search Recursive', y_name)
build_graph('levenshteinDistance.csv', 'Levenshtein Distance', y_name)

print('*********** TIME ***********')
y_name = 'time'
build_graph('study1.csv', 'Study 1', y_name)
build_graph('study2.csv', 'Study 2', y_name)
build_graph('shell.csv', 'Shell', y_name)
build_graph('selection_sort.csv', 'Selection Sort', y_name)
build_graph('quick_sort.csv', 'Quick Sort', y_name)
build_graph('heapsort.csv', 'Heapsort', y_name)
build_graph('insertion_sort.csv', 'Insertion Sort', y_name)
build_graph('insertion_sort_bin.csv', 'Insertion Sort Bin', y_name)
build_graph('binary_search.csv', 'Binary Search', y_name)
build_graph('binary_search_recursive.csv', 'Binary Search Recursive', y_name)
build_graph('levenshteinDistance.csv', 'Levenshtein Distance', y_name)
```
