```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def build_graph(algorithm, title, y):
    df = pd.read_csv(algorithm)

    # Criando uma figura com uma grade de subplots
    fig, axs = plt.subplots(1, 2, figsize=(8, 4))

    # Plotando o primeiro gráfico no primeiro subplot
    axs[0].scatter(df['n'], df[y])
    axs[0].set_ylabel(y)
    axs[0].set_xlabel('n')
    axs[0].set_title(title + ' Algorithm')

    # Plotando o segundo gráfico no segundo subplot
    axs[1] = fig.add_subplot(122, projection='3d')
    axs[1].scatter(df['n'], df[y], df['code_line'])
    axs[1].set_xlabel('n')
    axs[1].set_ylabel(y)
    axs[1].set_zlabel('code line')
    axs[1].set_title(title + ' Algorithm')

    # Ajustando o layout
    plt.tight_layout()
    plt.show()
```

```python
print('*********** HITS ***********')

y_name = 'hits'
build_graph('study1.csv', 'Study 1', y_name)
build_graph('study2.csv', 'Study 2', y_name)
build_graph('shell.csv', 'Shell', y_name)
build_graph('selection_sort.csv', 'Selection Sort', y_name)
build_graph('quick_sort.csv', 'Quick Sort', y_name)
build_graph('heapsort.csv', 'Heapsort', y_name)
build_graph('insertion_sort.csv', 'Insertion Sort', y_name)
build_graph('insertion_sort_bin.csv', 'Insertion Sort Bin', y_name)
build_graph('binary_search.csv', 'Binary Search', y_name)
build_graph('binary_search_recursive.csv', 'Binary Search Recursive', y_name)
build_graph('levenshteinDistance.csv', 'Levenshtein Distance', y_name)

print('*********** TIME ***********')
y_name = 'time'
build_graph('study1.csv', 'Study 1', y_name)
build_graph('study2.csv', 'Study 2', y_name)
build_graph('shell.csv', 'Shell', y_name)
build_graph('selection_sort.csv', 'Selection Sort', y_name)
build_graph('quick_sort.csv', 'Quick Sort', y_name)
build_graph('heapsort.csv', 'Heapsort', y_name)
build_graph('insertion_sort.csv', 'Insertion Sort', y_name)
build_graph('insertion_sort_bin.csv', 'Insertion Sort Bin', y_name)
build_graph('binary_search.csv', 'Binary Search', y_name)
build_graph('binary_search_recursive.csv', 'Binary Search Recursive', y_name)
build_graph('levenshteinDistance.csv', 'Levenshtein Distance', y_name)
```
